## Manipulación de arreglos numéricos

Si ha usado python es posible que esté habituado a organizar arreglos de variables en listas. Si bien son estructuras útiles para propósitos generales, no son ideales para la cálculo numérico.

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------
# Operaciones básicas con listas
# -----------------------
x = [1, 2, 3]

# Multiplicación escalar usando comprensión de listas
y = [5 * i for i in x]
print("y (list comprehension):", y)

# Multiplicación escalar usando map y lambda
y = list(map(lambda i: 5 * i, x))
print("y (map + lambda):", y)

# -----------------------
# Operaciones con arrays de NumPy
# -----------------------
x = np.array([1, 3, 5])
print("x (array 1D):", x)

y = 5 * x
print("y = 5 * x:", y)

# Array 2D
x = np.array([[1, 2, 3],
              [-5, -8, 0],
              [3, 0, 1]])
print("x (array 2D):\n", x)

y = np.array([[-2, 0, 2],
              [-1, -1, 0],
              [0, 2, 3]])
print("y (array 2D):\n", y)

# Multiplicación elemento a elemento
z = x * y
print("Multiplicación elemento a elemento (x*y):\n", z)

# Multiplicación matricial
z_mat = x @ y
print("Multiplicación matricial (x@y):\n", z_mat)

# Operaciones element-wise (ejemplo: tangente hiperbólica)
z_tanh = np.tanh(x)
print("tanh(x):\n", z_tanh)

# -----------------------
# Indexación y slicing
# -----------------------
print("Elemento z[1,1]:", z_tanh[1, 1])
print("Fila 0 completa:", z_tanh[0, :])
print("Fila 0, columnas >= 1:", z_tanh[0, 1:])

# -----------------------
# Álgebra lineal
# -----------------------
# Autovalores
autovals = np.linalg.eigvals(z_tanh)
print("Autovalores de z_tanh:", autovals)

# Matriz inversa
try:
    inversa = np.linalg.inv(z_tanh)
    print("Matriz inversa de z_tanh:\n", inversa)
    # Comprobación de la inversa
    print("z_tanh @ inversa (debería ser identidad):\n", z_tanh @ inversa)
except np.linalg.LinAlgError:
    print("La matriz no es invertible.")
```

## Resolver ecuaciones diferenciales

Las Ecuaciones Diferenciales Ordinarias (EDO), son aquellas en las que hay una úncia variable de integración (variable independiente).

Empecemos con un ejemplo más sencillo de la forma

$$
\frac{dx}{dt}=f(x,t)
$$

La idea es calcular la función $x(t)$ que cumple la ecuación diferencial arriba especificada. Para poder calcular la solución $x(t)$ de la ecuación anterior es necesario especificar una condición inicial (IVP), es decir, debemos conocer el valor de $x$ en un cierto instante, por ejemplo en $t=0$.

Veamos algunos métodos de solución numérica de EDO (integración).


### Método de Euler

La expansión en series de Taylor de la función $x(t+h)$, es decir nuestra función en un valor de $t$ conocido $x(t)$ más un intervalo pequeño $h$ es

$$
x(t+h)=x(t)+h\frac{dx}{dt}+\frac{1}{2}h^2 \frac{d^2x}{dt^2}+...
$$

$$
x(t+h)=x(t) + h f(x,t) + O(h^2)
$$

Nótese que para $h$ pequeño, los términos $h^2$ y de orden superior son negligibles. Este es el fundamento del método de Euler para EDO. La representación gráfica del método de Euler es la siguiente:

En caso de que tengamos varias ecuaciones diferenciales acopladas, podemos representar
la ecuación diferencial como:

$$
\frac{d\mathbf{x}}{dt}=\mathbf{f}(\mathbf{x},t)
$$

ahora $\mathbf{x} = (x_1(t),x_2(t),...,x_n(t))$ y $\mathbf{f}(\mathbf{x},t) = (f_1(\mathbf{x},t),f_2(\mathbf{x},t),...,f_n(\mathbf{x},t))$ y la aplicación del método de Euler es igual al caso de una sola EDO, i.e:

$$
\mathbf{x}(t+h)=\mathbf{x}(t)+h\frac{d\mathbf{x}}{dt}+\frac{1}{2}h^2 \frac{d^2\mathbf{x}}{dt^2}+...
$$

$$
\mathbf{x}(t+h)=\mathbf{x}(t) + h \mathbf{f}(\mathbf{x},t) + O(h^2)
$$

Considere el sistema de ecuaciones

$$
\frac{dV}{dt} = V-V^3 -w + I \\
\frac{dw}{dt} = \frac{1}{\tau}(V+a-bw)
$$

Asuma que $I = 1$, $a = 0.7$, $b = 0.8$, $\tau = 12.5$. Simule el sistema de ecuaciones durante 200 unidades de tiempo, sujeto a $(V(0),w(0)) = (0,0)$

```python
import matplotlib.pyplot as plt

# --- Parámetros del sistema ---
I = 1        # Corriente de entrada
a = 0.7
b = 0.8
tau = 12.5

# --- Condiciones iniciales ---
V = [0]      # Lista para almacenar V(t)
w = [0]      # Lista para almacenar w(t)
t = [0]      # Lista para almacenar el tiempo

# --- Parámetros de simulación ---
t_final = 200   # Tiempo final de la simulación
h = 0.01        # Tamaño del paso
N = int(t_final / h)  # Número de pasos

# --- Método de Euler ---
for i in range(N):
    V_actual = V[-1]
    w_actual = w[-1]
    t_actual = t[-1]

    # Cálculo de las derivadas
    dV_dt = V_actual - V_actual**3 - w_actual + I
    dw_dt = (V_actual + a - b*w_actual) / tau

    # Actualización usando Euler
    V_new = V_actual + h * dV_dt
    w_new = w_actual + h * dw_dt

    # Guardar valores en las listas
    V.append(V_new)
    w.append(w_new)
    t.append(t_actual + h)

# --- Graficar los resultados ---
plt.figure(figsize=(10,5))
plt.plot(t, V, label='V(t)')
plt.plot(t, w, label='w(t)')
plt.xlabel('Tiempo')
plt.ylabel('Variables')
plt.title('Simulación del sistema usando Euler')
plt.legend()
plt.grid(True)
plt.show()
```

## Método de Heun (Euler mejorado):

```python
import matplotlib.pyplot as plt

# Parámetros
I, a, b, tau = 1, 0.7, 0.8, 12.5
t_final, h = 200, 0.01
N = int(t_final/h)

# Inicialización
V, w, t = [0], [0], [0]

# Método de Heun
for i in range(N):
    Vn, wn = V[-1], w[-1]

    # Pendiente inicial (Euler)
    dV_dt = Vn - Vn**3 - wn + I
    dw_dt = (Vn + a - b*wn)/tau

    # Predicción Euler
    V_star = Vn + h * dV_dt
    w_star = wn + h * dw_dt

    # Pendiente corregida
    dV_dt_star = V_star - V_star**3 - w_star + I
    dw_dt_star = (V_star + a - b*w_star)/tau

    # Actualización final
    V_new = Vn + h/2 * (dV_dt + dV_dt_star)
    w_new = wn + h/2 * (dw_dt + dw_dt_star)

    V.append(V_new)
    w.append(w_new)
    t.append(t[-1] + h)

# Graficar
plt.figure(figsize=(10,5))
plt.plot(t, V, label='V(t) Heun')
plt.plot(t, w, label='w(t) Heun')
plt.xlabel('Tiempo')
plt.ylabel('Voltaje')
plt.legend()
plt.grid(True)
plt.show()
```

# Método más eficiente

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# --- Parámetros del sistema ---
I = 1       # Corriente de entrada
a = 0.7
b = 0.8
tau = 12.5

# --- Definición del sistema FitzHugh-Nagumo ---
def fitzhugh_nagumo(t, x):
    """
    Sistema de ecuaciones diferenciales de FitzHugh-Nagumo:

    x[0] = V(t)  : potencial de membrana
    x[1] = w(t)  : variable de recuperación

    Devuelve [dV/dt, dw/dt].
    """
    V, w = x
    dV_dt = V - V**3 - w + I
    dw_dt = (V + a - b*w) / tau
    return [dV_dt, dw_dt]

# --- Condiciones iniciales ---
x0 = [0, 0]  # [V(0), w(0)]

# --- Intervalo de integración ---
t_span = (0, 200)  # Desde t=0 hasta t=200
t_eval = np.linspace(t_span[0], t_span[1], 20000)  # Puntos donde queremos la solución

# --- Resolución del sistema usando solve_ivp ---
sol = solve_ivp(
    fun=fitzhugh_nagumo,  # Sistema de ecuaciones
    t_span=t_span,        # Intervalo de tiempo
    y0=x0,                # Condiciones iniciales
    t_eval=t_eval,        # Puntos donde evaluamos la solución
    method='RK45'         # Método Runge-Kutta de orden 5(4)
)

# --- Graficar resultados ---
plt.figure(figsize=(10,5))
plt.plot(sol.t, sol.y[0], label='V(t)')
plt.plot(sol.t, sol.y[1], label='w(t)')
plt.xlabel('Tiempo')
plt.ylabel('Voltaje')
plt.title('Simulación del modelo FitzHugh-Nagumo')
plt.legend()
plt.grid(True)
plt.show()
```

## Postprocesar resultados

En nuestro periplo por las neurociencias computacoinales veremos la importancia de identificar los momentos en los que la variable que describe el comportamiento de una neurona tiene picos, llamados *spikes* o potenciales de acción. Veamos cómo identificar picos

```python
V = sol.y[0]
t = t_eval

from scipy.signal import find_peaks


peackId, _ = find_peaks(V)
tiempos_spike = t[peackId]
plt.plot(t, V)
plt.plot(tiempos_spike, V[peackId], 'o')
plt.xlabel('Tiempo')
plt.ylabel('Voltaje')
plt.show()
```
